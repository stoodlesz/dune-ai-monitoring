```python
import requests
import os

url = "https://github.com/mapbox/rasterio/raw/main/tests/data/RGB.byte.tif"
file_path = "data/raw/sample_satellite.tif"

os.makedirs("data/raw", exist_ok=True)

print("Downloading sample GeoTIFF...")

response = requests.get(url)
response.raise_for_status()

with open(file_path, "wb") as f:
    f.write(response.content)

print("Download complete:", file_path)
```

```python
import rasterio

with rasterio.open("data/raw/sample_satellite.tif") as src:
    print(src.count)
    print(src.width, src.height)
    print(src.crs)
```

```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

with rasterio.open("data/raw/sample_satellite.tif") as dataset:
    r = dataset.read(1)
    g = dataset.read(2)
    b = dataset.read(3)

rgb = np.dstack((r, g, b))

plt.imshow(rgb)
plt.title("Satellite Image Sample (RGB)")
plt.axis("off")
plt.show()
```
